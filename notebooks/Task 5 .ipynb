```python
# Define asset tickers
tickers = ["TSLA", "BND", "SPY"]

# Define date range
start_date = "2015-01-01"
end_date = "2026-06-30"
```

```python
# Download data into separate DataFrames
tsla_df = yf.download("TSLA", start=start_date, end=end_date)
bnd_df = yf.download("BND", start=start_date, end=end_date)
spy_df = yf.download("SPY", start=start_date, end=end_date)

# Display the first few rows
print("TSLA")
print(tsla_df.head())

print("\nBND")
print(bnd_df.head())

print("\nSPY")
print(spy_df.head())
```

```python
# Download data for all assets
data = yf.download(
    tickers,
    start=start_date,
    end=end_date,
    group_by="ticker",
    auto_adjust=False
)

print(data.head())
```

```python
combined_data = []

for ticker in tickers:
    df = yf.download(
        ticker,
        start=start_date,
        end=end_date,
        auto_adjust=False
    )

    df = df.reset_index()
    df["Asset"] = ticker

    combined_data.append(df)

# Combine all assets
historical_data = pd.concat(combined_data, ignore_index=True)

# View the data
print(historical_data.head())
```

```python
# Save combined dataset
historical_data.to_csv("historical_financial_data.csv", index=False)

print("Data saved successfully.")
```

```python
import numpy as np

def total_return(cumulative_value):
    return cumulative_value.iloc[-1] - 1

def annualized_return(daily_returns):
    n = len(daily_returns)
    return (1 + daily_returns).prod() ** (252 / n) - 1

def sharpe_ratio(daily_returns, risk_free_rate=0.0):
    excess = daily_returns - risk_free_rate / 252
    return np.sqrt(252) * excess.mean() / excess.std()

def max_drawdown(cumulative_value):
    running_max = cumulative_value.cummax()
    drawdown = (cumulative_value - running_max) / running_max
    return drawdown.min()
```

```python
portfolio_metrics = {
    "Total Return": total_return(portfolio_value),
    "Annualized Return": annualized_return(portfolio_returns),
    "Sharpe Ratio": sharpe_ratio(portfolio_returns),
    "Maximum Drawdown": max_drawdown(portfolio_value)
}

benchmark_metrics = {
    "Total Return": total_return(benchmark_value),
    "Annualized Return": annualized_return(benchmark_returns),
    "Sharpe Ratio": sharpe_ratio(benchmark_returns),
    "Maximum Drawdown": max_drawdown(benchmark_value)
}
```

```python
import pandas as pd

results = pd.DataFrame({
    "Optimized Portfolio": portfolio_metrics,
    "60/40 Benchmark": benchmark_metrics
})

# Convert returns to percentages
results.loc["Total Return"] *= 100
results.loc["Annualized Return"] *= 100
results.loc["Maximum Drawdown"] *= 100

print(results.round(2))
```
