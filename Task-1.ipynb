```python
import pandas as pd
import yfinance as yf

tickers = ["TSLA", "BND", "SPY"]

frames = []

for ticker in tickers:

    df = yf.Ticker(ticker).history(
        start="2015-01-01",
        end="2025-12-31",
        auto_adjust=False
    )

    df.reset_index(inplace=True)

    df["Asset"] = ticker

    frames.append(df)

combined_df = pd.concat(
    frames,
    ignore_index=True
)

print(combined_df.head())
```
